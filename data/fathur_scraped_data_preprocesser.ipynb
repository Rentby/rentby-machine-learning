```python
import pandas as pd
import os
```

```python
data_dir = 'fathur_new_scraped_desc/'
hiking_dir = 'hiking/'
cosplay_dir = 'cosplay/'
```

```python
def get_dataframe(data_dir, hobby_dir):
    path = os.path.join(data_dir, hobby_dir)
    files = os.listdir(path)
    data = []
    for file in files:
        df = pd.read_csv(os.path.join(path, file))
        data.append(df)
    return pd.concat(data, ignore_index=True)
```

```python
hiking_df = get_dataframe(data_dir, hiking_dir)
hiking_df
```

```python
cosplay_df = get_dataframe(data_dir, cosplay_dir)
cosplay_df
```

```python
hiking_df['cosplay'] = 0
cosplay_df['cosplay'] = 1

hiking_df['hiking'] = 1
cosplay_df['hiking'] = 0
```

```python
all_df = pd.concat([hiking_df, cosplay_df], ignore_index=True)
all_df.drop_duplicates(subset=['product_name'], inplace = True)
all_df.dropna(inplace = True)
all_df.reset_index(drop=True, inplace=True)
all_df
```

```python
all_df.rename(columns = {'seller': 'renters', 'img_url': 'url_photo', 'desc': 'description'}, inplace=True)
all_df
```

```python
new_order = ['product_name', 'rent_price', 'url_photo', 'link', 'renters', 'description', 'cosplay', 'hiking']
all_df = all_df[new_order]
all_df
```

```python
def clean_and_convert_price(price_str):
    clean_str = price_str.replace('Rp', '').replace('.', '')
    rent_price = int(clean_str)/10
    return int(rent_price)

# Apply the function to the rent_price column
all_df['rent_price'] = all_df['rent_price'].apply(clean_and_convert_price)

# Show rent price
all_df
```

```python
all_df.to_csv('fathur_new_data.csv', index=False)
```
